```python
import matplotlib.patches as mpatches
fig, ax = plt.subplots(figsize=(8,5))
exposed_idx = np.arange(3)
bar_width = 0.3
learned_idx = exposed_idx + bar_width
exposed_bar=[89.47,90.00,77.55]
learned_bar=[88.72,84.12,79.59]
p1 = ax.bar(exposed_idx, height=exposed_bar, width=bar_width, color = ["yellow","green","red"],label="Exposed")
p2 = ax.bar(learned_idx, height=learned_bar, width=bar_width, color = ["yellow","green","red"],label="Learned",hatch="\\\\")
ax.set_ylabel("Accuracy (%)")
ax.set(ylim=(0, 119))
ax.set_xticks(learned_idx-0.15)
ax.set_xticklabels(('Control', 'Difficulty', 'Random'))
h1 = mpatches.Patch(facecolor="lightgrey",edgecolor="black",lw=1,label="Without color cues")
h2 = mpatches.Patch(facecolor="lightgrey",edgecolor="black",lw=1,label="With color cues (if applicable)",hatch="\\\\")
plt.legend(handles=[h1,h2], loc='upper left')
plt.title("Accuracy per group (testing phase)")
plt.tight_layout()
# plt.savefig("fig_acc_cond.png")
plt.show()
```
